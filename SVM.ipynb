### Support Vector Machines

```python
import numpy as np
import sklearn as sk
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data=np.genfromtxt("linearlyseparable.csv")
```

data.shape

```python
X=data[:,range(2)]
```

```python
y=data[:,2]
```

```python
plt.scatter(X[:,0], X[:,1],c=y,s=70,cmap='winter')
```

***Now I will load the SVC package***

```python
from sklearn.svm import SVC
```

***In the SVC constructor below, we set the kernel parameter to 'linear', because we know, from the shape of data that a linear separation boundary exists. The C parameter is used to allow for soft-margins, meaning some errors in classification is tolerable. High C means no tolerance for mis-classification.***

```python
clf=SVC(kernel='linear', C=1E10)
```

```python
clf.fit(X,y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
np.sum(clf.predict(X)!=y)
```

***Plotting the decision boundary***

```python
h=.01
x_min=-1
x_max=4
y_min=-1
y_max=6
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z) #play around with cmap options to get different color maps
plt.scatter(X[:,0], X[:,1],c=y,s=70,cmap='winter')
plt.show()
```

***Using Contourf()***

```python
h=.01
x_min=-1
x_max=4
y_min=-1
y_max=6
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z,cmap=plt.cm.coolwarm, alpha=0.7) #play around with cmap options to get different color maps
plt.scatter(X[:,0], X[:,1],c=y,s=70,cmap='winter')
plt.show()
```
